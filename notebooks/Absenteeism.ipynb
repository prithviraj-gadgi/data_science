```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('./../data/Absenteeism_data.csv')
```

```python
raw_data.head()
```

```python
raw_data.shape
```

```python
raw_data.info()
```

```python
raw_data.describe(include='all')
```

```python
raw_data.drop(['ID'], axis=1, inplace=True)
```

```python
raw_data.head()
```

```python
sorted(raw_data['Reason for Absence'].unique())
```

```python
reason_columns = pd.get_dummies(raw_data['Reason for Absence'], drop_first=True)
```

```python
reason_columns.head()
```

```python
reason_columns = reason_columns.astype(int)
```

```python
reason_columns.head()
```

```python
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)
```

```python
raw_data = pd.concat([raw_data, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)
```

```python
raw_data.head()
```

```python
raw_data.drop(['Reason for Absence'], axis=1, inplace=True)
```

```python
raw_data.head()
```

```python
raw_data.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
raw_data.columns = column_names
```

```python
raw_data.head()
```

```python
raw_data.columns.values
```

```python
reorderd_columns = ['Reason_1',
       'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
raw_data = raw_data[reorderd_columns]
```

```python
raw_data.head()
```

```python
type(raw_data['Date'][0])
```

```python
raw_data['Date'] = pd.to_datetime(raw_data['Date'], format='%d/%m/%Y')
```

```python
type(raw_data['Date'][0])
```

```python
raw_data['Date'][0].month
```

```python
list_months = []
for i in range(len(raw_data['Date'])):
    list_months.append(raw_data['Date'][i].month)
```

```python
raw_data['Month Value'] = list_months
raw_data.head()
```

```python
raw_data['Date'][0].weekday()
```

```python
def date_to_weekday(date):
    return date.weekday()
```

```python
raw_data['Day of the week'] = raw_data['Date'].apply(date_to_weekday)
raw_data.head()
```

```python
raw_data.drop(['Date'], axis=1, inplace=True)
```

```python
raw_data.columns.values
```

```python
col_list = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value',
       'Day of the week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
raw_data = raw_data[col_list]
raw_data.head()
```

```python
raw_data.info()
```

```python
raw_data['Education'].unique()
```

```python
raw_data['Education'].value_counts()
```

```python
education_col = pd.get_dummies(raw_data['Education'], drop_first=True)
```

```python
education_col.head()
```

```python
raw_data = pd.concat([raw_data, education_col], axis=1)
raw_data.head()
```
