```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
a = np.arange(1,100)
a
```

```python
b = np.arange(101, 200)
b
```

```python
a_train, a_test = train_test_split(a)
a_train, a_test
```

```python
b_train, b_test = train_test_split(b)
b_train, b_test
```

```python
a_train, a_test, b_train, b_test = train_test_split(a, b, random_state=42, test_size=0.2)
```
