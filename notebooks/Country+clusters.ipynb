```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('./../data/3.01.+Country+clusters.csv')
```

```python
data
```

```python
plt.scatter(data['Longitude'], data['Latitude'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```

```python
x = data.iloc[:, 1:3]
x
```

```python
kmeans = KMeans(3)
kmeans.fit(x)
```

```python
clusters = kmeans.fit_predict(x)
clusters
```

```python
plt.scatter(data['Longitude'], data['Latitude'], c=clusters, cmap='rainbow')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```
