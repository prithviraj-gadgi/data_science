```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'A': [1,2,3,1,2,3],
    'B': [1,1,1,2,2,2],
    'C': [5,6,7,8,9,10]
})
```

```python
pd.pivot_table(df, index='A', columns='B', values='C', aggfunc='sum')
```

```python
df.groupby(['A', 'B'])['C'].sum()
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
sns.boxplot(data=df, x="species", y="sepal_length")
```

```python
sns.boxplot(data=df, x="species", y="sepal_length")
sns.despine()
```

```python
g = sns.FacetGrid(df, col="species", height=3)
g.map(sns.scatterplot, "sepal_length", "sepal_width")
```

```python
import tensorflow as tf
tf.random.set_seed(0)
a = tf.constant([1,2,3])
b = tf.constant([4,5,0])
print(a + b)
```

```python
@tf.function
def square(x):
    return x * x

x = tf.constant([2,3])
print(square(x))
```
